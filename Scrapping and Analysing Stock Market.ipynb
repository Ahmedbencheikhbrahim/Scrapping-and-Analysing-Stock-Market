```python
volatilite = echantillon.pct_change().std() * np.sqrt(252)
```

```python
 rendement["volatilite"] = volatilite
```

```python
 rendement
```

```python
data = np.asarray([np.asarray(rendement['rendement']),np.asarray(rendement['volatilite'])]).T
```

```python
X = data
inertie =[]
for k in range(2,10):
  k_means = KMeans(n_clusters = k)
  k_means.fit(X)
  inertie.append(k_means.inertia_)
```

```python
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 10), inertie)
plt.xticks(range(2,10))
plt.grid(True)
plt.title('Elbow curve');
```
